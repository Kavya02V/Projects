## Data Visualization

#### scatterplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='low', y='high', hue='high', data=data)
plt.title('low vs. high (Seaborn)')
plt.show()
```

#### Matplotlib

```python
import matplotlib.pyplot as plt

plt.plot(data['low'], data['high'], marker='o', label='Stock market prediction')
plt.xlabel('low')
plt.ylabel('high')
plt.title('Stock market prediction (Matplotlib)')
plt.legend()
plt.grid()
plt.show()
```

#### histplot

```python
sns.histplot(data["low"], color="brown")
plt.title("low")
plt.show()
```

#### boxplot

```python
sns.boxplot(y=data["high"], color="orange")
plt.title("high")
plt.show()
```

#### pairplot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\Kavya Venkatesh\Downloads\archive (15)\infolimpioavanzadoTarget.csv")  

print(data.columns)

sns.pairplot(data, vars=['open', 'high', 'low',	'close', 'adjclose', 'volume'], hue='high', palette='coolwarm')

plt.suptitle('Pair Plots of Selected Features')  
plt.show()
```

#### Volume Charts 


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load stock data (example: Apple stock)
data = pd.read_csv(r"C:\Users\Kavya Venkatesh\Downloads\archive (15)\infolimpioavanzadoTarget.csv", parse_dates=['close'])
data.set_index('close', inplace=True)

# Plot volume chart
plt.figure(figsize=(12, 6))
plt.bar(data.index, data['high'], color='purple', alpha=0.7)
plt.title('Stock Trading Volume Over Time')
plt.xlabel('close')
plt.ylabel('high')
plt.show()
```

```python
data.groupby('low')['high'].sum().plot(kind='line', color='purple')
plt.xlabel('low')
plt.ylabel('high')
plt.title('Stock Trading')
plt.show()
```
